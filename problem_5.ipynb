```python
src_user_time_raw = pd.Series(df.src_user.values,index=df._time).to_dict()
src_user_time = defaultdict(list)
```

```python
for k, v in src_user_time_raw.items():
    src_user_time[v].append(dp.parse(k).strftime('%s'))
```

```python
def find_regularities(d):
    res = dict()
    for user, time_list in d.items():
        time_list_s = sorted(time_list)
        diff = list()
        for i in range(len(time_list_s) - 1):
            diff.append(int(time_list_s[i+1]) - int(time_list_s[i]))
        diff = set(diff)
        if (len(diff) == 1):
            res[user] = diff.pop()
    return res
```

```python
res_d = find_regularities(src_user_time)
```
